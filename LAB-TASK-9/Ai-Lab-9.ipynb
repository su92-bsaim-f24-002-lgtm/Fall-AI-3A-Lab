1. Read CSV:

```python
import pandas as pd

dataset = pd.read_csv("titanic_big.csv")  
```

```python

# 7. Print datatype of each column
print("\nData types of columns:")
print(dataset.dtypes)
```

2. Print Top 5 rows of dataset:

```python
print("Top 5 rows of Dataset:")
print(dataset.head())
```

3. Print Bottom 5 rows

```python
print("\nBottom 5 rows of Dataset:")
print(dataset.tail())
```

4.Print Number of Rows and Columns

```python
print("\nNumber of Rows and Columns:")
print("Rows: ", dataset.shape[0])
print("Columns: ", dataset.shape[1])
```

5. Print/Determine all null columns

```python
print("\nTotal null/missing values in Dataset:")
print(dataset.isnull().sum())
```

6. Filling null values using Mean / Median / Mode

```python
dataset = dataset.fillna(dataset.mean(numeric_only=True))
```

```python
dataset = dataset.fillna(dataset.mode().iloc[0])
```

```python
print("\nNull values after filling:")
print(dataset.isnull().sum())
```
